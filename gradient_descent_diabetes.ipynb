```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import array as ar
import pandas as pd

df = pd.read_csv('Diabetes.csv')
target = df['Target']
cholesterol = df['Cholesterol']
BMI = df['BMI']
BP = df['BP']
LDL = df['LDL']

x1g, x2g, x3g, yg = cholesterol, BMI, BP, target

def prediction(x1, a, b):
    return a * x1 + b
```

```python
# stochastic gradient descent 
lr = 0.00001
sample_num = len(yg)
a, b = 0, 0
rs = []
rmse = np.inf
for i in range(10):
    plt.scatter(x1g, yg, color='blue')
    plt.plot(x1g, prediction(x1g, a, b), color='red')
    plt.xlabel('cholesterol')
    plt.ylabel('target (degree of diabetes)')
    plt.show()            
    rmse_before = rmse
    rmse = np.sqrt(((yg - prediction(x1g, a, b)) ** 2).mean())
    rs.append(rmse)
    if np.abs(rmse_before - rmse) < 0.001:
        break
    error = yg - prediction(x1g, a, b)
    a += lr * np.sum(error * x1g) / sample_num
    b += lr * np.sum(error) / sample_num
plt.plot(rs)
plt.show()
print('RMSE =', rmse)
```
