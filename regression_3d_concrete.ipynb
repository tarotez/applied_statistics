```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import array as ar
import pandas as pd
import matplotlib.cm as cm

df = pd.read_csv('concrete_data.csv')

strength = df['Strength']
cement = df['Cement']
water = df['Water']
sp = df['Superplasticizer']
fineness = df['Fine Aggregate']
age = df['Age']
```

```python
plt.scatter(cement, strength)
plt.show()
```

```python
x1r, x2r, yr = cement, age, strength

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x1r, x2r, yr, cmap=cm.coolwarm, linewidth=0, antialiased=False, alpha=0.5)
plt.show()
```

```python
# visualize initial plane
x1r, x2r, yr = cement, water, strength
a, b, c = np.random.random(3)

def prediction(x1, x2, a, b, c):
    return a * x1 + b * x2 + c

x1s = np.linspace(min(x1r), max(x1r))
x2s = np.linspace(min(x2r), max(x2r))

x1g, x2g = np.meshgrid(x1s, x2s)
yg = ar([[prediction(x1, x2, a, b, c) for x1 in x1s] for x2 in x2s])

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x1r, x2r, yr, cmap=cm.coolwarm, linewidth=0, antialiased=False, alpha=0.5)
ax.plot_surface(x1g, x2g, yg, cmap=cm.coolwarm, linewidth=0, antialiased=False, alpha=0.5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
plt.show()

y_pred = [prediction(x1, x2, a, b, c) for x1, x2 in zip(x1r, x2r)]
rmse = np.sqrt(((yr - y_pred) ** 2).mean())
print('RMSE =', rmse)
```

```python
# gradient descent
x1r, x2r, yr = cement, water, strength

lr = 0.00001
epochNum = 1000

def prediction(x1, x2, a, b, c):
    return a * x1 + b * x2 + c

a, b, c = np.random.random(3)

for _ in range(epochNum):
    for x1, x2, y in zip(x1r, x2r, yr):
        error = y - prediction(x1, x2, a, b, c)
        a += lr * error * x1
        b += lr * error * x2
        c += lr * error

x1s = np.linspace(min(x1r), max(x1r))
x2s = np.linspace(min(x2r), max(x2r))

x1g, x2g = np.meshgrid(x1s, x2s)
yg = ar([[prediction(x1, x2, a, b, c) for x1 in x1s] for x2 in x2s])

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x1r, x2r, yr, cmap=cm.coolwarm, linewidth=0, antialiased=False, alpha=0.5)
ax.plot_surface(x1g, x2g, yg, cmap=cm.coolwarm, linewidth=0, antialiased=False, alpha=0.5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
plt.show()

y_pred = [prediction(x1, x2, a, b, c) for x1, x2 in zip(x1r, x2r)]
rmse = np.sqrt(((yr - ar(y_pred).reshape(yr.shape)) ** 2).mean())
print('RMSE =', rmse)
```

```python
# gradient descent
x1r, x2r, yr = cement, water, strength
X = ar([x1r, x2r, np.ones(len(x1r))]).T
y = ar(yr).reshape(-1, 1)
lr = 1e-8
epochNum = 100

def prediction(X, w):
    return X @ w

w = np.random.random(X.shape[1]).reshape(-1, 1)
for _ in range(epochNum):
    # error = y - prediction(X, w)
    error = y - X @ w
    w += lr * (error.T @ X).T

y_pred = prediction(X, w)
rmse = np.sqrt(((yr - ar(y_pred).reshape(yr.shape)) ** 2).mean())
print('RMSE =', rmse)
```
