```python
# gradient descent
from numpy import array as ar
import pandas as pd
df = pd.read_csv('concrete_data.csv')
strength = df['Strength']
cement = df['Cement']
water = df['Water']
x1r, x2r, yr = cement, water, strength
X = ar([x1r, x2r, np.ones(len(x1r))]).T
y = ar(yr).reshape(-1, 1)

lr = 1e-8
epochNum = 10
trialNum = 10

def prediction(X, w):
    return X @ w

rmse_L = []
for _ in range(trialNum):
    w = np.random.random(X.shape[1]).reshape(-1, 1)
    for epochID in range(epochNum):
        error = y - prediction(X, w)
        w += lr * (error.T @ X).T

    y_pred = prediction(X, w)
    rmse = np.sqrt(((y - y_pred) ** 2).mean())
    print('RMSE =', rmse)
    rmse_L.append(rmse)
```

```python
from numpy import array as ar
import pandas as pd
df = pd.read_csv('concrete_data.csv')
strength = df['Strength']
cement = df['Cement']
water = df['Water']
x1r, x2r, yr = cement, water, strength
X = ar([x1r, x2r, x2r + x1r, np.ones(len(x1r))]).T
y = ar(yr).reshape(-1, 1)
# w = np.linalg.solve(X.T @ X, X.T @ y)
w = np.linalg.inv(X.T @ X) @ X.T @ y
y_pred = X @ w
rmse = np.sqrt(((y - y_pred) ** 2).mean())
print('RMSE =', rmse)
```
