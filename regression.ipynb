```python
import pandas as pd
import numpy as np
from numpy import array as ar
d = pd.read_csv("../data/GlobalTemperatures.csv")
```

```python
import matplotlib.pyplot as plt
offset = 36
plt.plot(d["LandAverageTemperature"][offset::12])
plt.show()
```

```python
import matplotlib.cm as cm
dates = d['dt'][offset::12]   # targets a specific month
years = [int(date.split('-')[0]) for date in dates]
temperatureSeq = np.array(d["LandAverageTemperature"][offset::])
temperatureMat = temperatureSeq.reshape(-1,12)
months = np.array(list(range(1,13)))
X, Y = np.meshgrid(months, years)
Z = np.array(temperatureMat)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
ax.set_xlabel('year')
ax.set_ylabel('month')
ax.set_zlabel('temperature')
plt.show()
```

```python
plt.plot(years, temperatureMat.mean(axis=1))
xs = years
a, b = 0, 0
ys = [a * x + b for x in xs]
plt.plot(xs, ys, color='red')
plt.xlabel('year')
plt.ylabel('temperature')
plt.show()
```
