```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import array as ar
import pandas as pd

df = pd.read_csv('Diabetes.csv')
target = df['Target']
cholesterol = df['Cholesterol']
BMI = df['BMI']
BP = df['BP']
LDL = df['LDL']

x1g, x2g, x3g, yg = cholesterol, BMI, BP, target

def prediction(x1, a, b):
    return a * x1 + b
```

```python
# stochastic gradient descent (batch size 1)
lr = 0.00001
a, b = 0, 0
rs = []
cnt = 0
rmse = np.inf
for i in range(1):
    for x1, y in zip(x1g, yg):
        if cnt % 1 == 0:
            plt.scatter(x1g, yg, color='blue')
            plt.plot(x1g, prediction(x1g, a, b), color='red')
            plt.xlabel('cholesterol')
            plt.ylabel('target (degree of diabetes)')
            plt.show()            
        rmse_before = rmse
        rmse = np.sqrt(((yg - prediction(x1g, a, b)) ** 2).mean())
        rs.append(rmse)
        if np.abs(rmse_before - rmse) < 0.001 or cnt > 20:
            break
        error = y - prediction(x1, a, b)
        a += lr * error * x1
        b += lr * error
        cnt += 1
plt.plot(rs)
plt.show()
```
